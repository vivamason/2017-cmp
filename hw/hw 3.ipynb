```python
import matplotlib.pyplot as plt
import numpy as np

#this changes the level of detail resolved. At this grid size there isnt much benefit past 2. 
detail = 2
#this changes the resolution of the image
grid_size = 300
#this changes the resolution of the algorithm. 
resolution = 50

def counter(c, cycles):
    #resets the value for the algorithm
    complex_num = complex(0, 0)
    #loops through the algorithm
    for n in range(cycles):
        complex_num = (complex_num*complex_num) + c
        #ends loop once the number is greater than the resolution desired. 
        if abs(complex_num) > detail:
            break
    return n

#declares the grid specifications. change the first and second value to change the size and scale of each axis
j = np.linspace(-2, 1, grid_size)
i = np.linspace(-2, 1, grid_size)

#declares the grid
grid = np.empty((len(j), len(i)))

#propogates the values for each pixel in the image
for x in range(len(j)):
    for y in range(len(i)):
        c = complex(j[x], i[y])
        grid[x, y] = counter(c, resolution)

#shows everything
plt.imshow(grid)
```
