```python
import matplotlib.pyplot as plt
import numpy as np
import time

def counter(c, cycles):
    #resets the value for the algorithm
    complex_num = complex(0, 0)
    #loops through the algorithm
    
    for n in range(cycles):
        complex_num = (complex_num*complex_num) + c
        #ends loop once the number is greater than the resolution desired. 
        if abs(complex_num) > detail:
            break
    return n



#propogates the values for each pixel in the image
def test(detail, grid_size, resolution):
    
    #declares the grid specifications. change the first and second value to change the size and scale of each axis
    j = np.linspace(-1, 1, grid_size)
    i = np.linspace(-1, 1, grid_size)

    #declares the grid
    grid = np.empty((len(j), len(i)))

    for x in range(len(j)):
        for y in range(len(i)):
            c = complex(j[x], i[y])
            grid[x, y] = counter(c, resolution)
            
value = []
seconds = []

#change this to change the number of trials done 
r=10


for u in range(r):
    start = time.time()
    
    grid_size_1 = u * 100
    test(2, grid_size_1, 80)
    value.append(grid_size_1)
    
    end = time.time()
    elapsed = end - start
    seconds.append(elapsed)

y = seconds
x = value
title = 'mandelbrot time scaling accross ' + str(r) + " trials"
plt.title(title)
plt.ylabel("seconds")
plt.xlabel("grid size value")
plt.plot(x,y)



    
```

```python
def Trapezoid(start, end, parts, plot=1):
    """
    Trapezoid sum rule
    
    """
    #define the function
    f = lambda x: np.exp(-x ** 2)  
    
    #define the X, Y points 
    deltax = (end - start) / parts
    resultsx = np.linspace(start, end, parts+1)
    resultsy = f(resultsx)

    # To calculate the area
    area = np.empty([parts], float)
    for i in range(parts):
            area[i] = (resultsy[i]+resultsy[i+1]) * deltax/2

    # By default, we also output the plot.
    if plot==1:
        x = np.linspace(start,end,100)
        y = f(x)        
        plt.plot(x, y, 'r')
        plt.xlim([start,end])
        
        y2 = np.array([0,0])
        for i in range(parts):
            x0 = resultsx[i:i+2]
            y1 = resultsy[i:i+2]
            plt.fill_between(x0, y1, y2, where=y1>=y2, facecolor='blue')
            linex, liney = [resultsx[i+1], resultsx[i+1]], [0, resultsy[i+1]]
            plt.plot(linex, liney, color='black', linewidth=2.0)        
        
        plt.show()
        print("The Sum of the area is: ", sum(area))
    
    return sum(area)

Trapezoid(0,3,5)
```
