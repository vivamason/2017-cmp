```python
#this computes the factorial
def factorial(B):
    z = B
    while (B>1):
        B = B - 1
        z = B * z
    return z

#this does the einstein solid computation
def compute(q, N):
    top_line = q + N -1
    top_line = factorial(top_line)
    temp1 = N - 1
    bottom_line = factorial(q) * factorial(temp1)
    result = top_line / bottom_line
    return result


#im seting the array for the first simulation
arr = []
#this cycles through the first 200 integers and stores them in an array 
for x in range(1,201):
    Q_a = compute(x,300)
    Q_b = compute(100,200)
    Q_total = Q_a * Q_b
    arr.append(Q_total)
    

import matplotlib.pyplot as plt
from numpy import *
import matplotlib.pyplot as plt

#this sets the y axis as the value of number in the array
y = arr
#this sets the x axis as the number of ingegers computed
x = range(len(arr))
#this sets the plot
ll0 = plt.plot(x,y)
#this makes the y axis logarithmic while keeping the x axis linear
plt.semilogy(x, y)
plt.title('as a function of N_A=300, N_B=200, and q=100')

plt.show()

#this is the same as the above code except with the different set of values
arr2 = []
for x in range(1,51):
    Q_a = compute(x,3000)
    Q_b = compute(100,2000)
    Q_total = Q_a * Q_b
    arr2.append(Q_total)

y = arr2
x = range(len(arr2))
ll = plt.plot(x,y)
plt.semilogy(x, y)
plt.title('as a function of N_A=3000, N_B=2000, and q=100')

plt.show()
```
